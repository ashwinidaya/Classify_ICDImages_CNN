```
from icrawler.builtin import  GoogleImageCrawler
from fastai.vision import *
from pathlib import Path
from random import randint
import logging
```

```
classes = ['Bharathanatyam','Kathakali','Kathak','Odissi','Yakshagana']
```

```
for c in classes:
    print(c)
    download_directory = './data/dances/'+c
    search_filters_google = dict()
    google_crawler = GoogleImageCrawler(downloader_threads=4,storage={'root_dir': download_directory})
    google_crawler.crawl(keyword=c, filters=search_filters_google, max_num=1000)
    verify_images(download_directory, delete=True, max_workers=8)
```

```
import numpy as np
from fastai.vision import *
from fastai import *
from pathlib import Path
```

```
#np.random.seed(8)
PATH= Path('./data/dances')
#path = Path('data/dances/'+classes[randint(0,3)])
data = ImageDataBunch.from_folder(PATH, valid_pct=0.2, train=".", size=224,ds_tfms=get_transforms(),num_workers=4).normalize(imagenet_stats)
```

```
data.classes
```

```
data.show_batch(rows=3, figsize=(7,8))
```

```
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```
learn = create_cnn(data=data,base_arch=models.resnet34, metrics=error_rate)
```

```
learn.fit_one_cycle(10)
```

```
learn.save('stage-1')
```

```
learn.unfreeze()
```

```
learn.lr_find()
```

```
#{learner_name}.recorder.plot()
learn.recorder.plot()
```

```
learn.fit_one_cycle(10, max_lr=slice(3e-7,1e-6))
```

```
learn.save('stage-2')
```

```

learn.load('stage-2')
```

```
interp = ClassificationInterpretation.from_learner(learn)
```

```
interp.plot_confusion_matrix()
```
